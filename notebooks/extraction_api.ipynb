```python
def get_playlist_details(spotify,playlist_id):
    playlist_details = []
    offset = 0
    while offset<10000:
        items = spotify.playlist_items(playlist_id,market='US',offset=offset)["items"]
        playlist_details = playlist_details + items
        offset+=100
    return playlist_details
```

```python
def get_track_details(spotify,track_ids):
    tracks = []
    artists = []
    for track_id in track_ids:
        print(track_id)
        track = spotify.track(track_id)
        track_dic = {}
        
        artists.append([artist['id'] for artist in track['artists']])
        
        track_dic['track_popularity'] = track['popularity']
        track_dic['album_released_date'] = track['album']['release_date']
        track_dic['id'] = track['id']
        tracks.append(track_dic)
    return (tracks,artists)
```

```python
def get_artist_details_for_track(spotify,artists_ids):
    genres = []
    average_popularity = []
    artists_batch = []
    group_counter = 0
    for artist_id_group in artists_ids:
        if len(artist_id_group) + len(artists_batch) <= 50:
            group_counter += 1
            artists_batch = artists_batch + artist_id_group
        else:
            print(len(artists_batch))
            artist_details = spotify.artists(artists_batch)
            offset = 0
            for group_index in range(group_counter):
                group_size = len(artists_ids[group_index])
                genres.append([artist['genres'] for artist in artist_details['artists'][offset:offset+group_size]])
                average_popularity.append(sum([artist['popularity'] for artist in artist_details['artists'][offset:offset+group_size]])/group_size)
                offset+=group_size
            artists_batch=[]
            group_counter = 0
    if artists_batch:
        artist_details = spotify.artists(artists_batch)
        offset = 0
    for group_index in range(group_counter):
        group_size = len(artists_ids[group_index])
        genres.append([artist['genres'] for artist in artist_details['artists'][offset:offset+group_size]])
        average_popularity.append(sum([artist['popularity'] for artist in artist_details['artists'][offset:offset+group_size]])/group_size)
        offset+=group_size
    return (genres,average_popularity)
```
